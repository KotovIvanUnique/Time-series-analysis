3. Создадим два АР(1) процесса первого порядка:

$z_t = 0.8*z_{t-1}+a_t$

$z_t = -0.8*z_{t-1}+a_t$

где $a_t$ – случайная нормально распределенная величина малой
амплитуды (порядка 0.2), $z_0=1$

```python
z1 = np.zeros(100)
z2 = np.zeros(100)
z1[0] = 1
z2[0] = 1
for i in range(1,100):
    z1[i] = 0.8 * z1[i - 1] + 0.2 * np.random.randn()
    z2[i] = -0.8 * z2[i - 1] + 0.2 * np.random.randn()
plt.figure(figsize = (10, 5))
plt.plot(z1, 'b')
plt.plot(z2, 'r')
plt.show()
```

4. Постройте для этих рядов функции автокорреляции с помощью функции plot_acf:

```python
plt.figure(figsize = (10, 5))
plot_acf(z1, lags=50)
plot_acf(z2, lags=50)
plt.show()
```
